Optimization model: Profit Maximization for a Manufacturig Company

Objective: Maximize the profit from producing Products A and B.

Decision Variable:
x : Number of units of Product A.
y : Number of units of Product B.

Contraints: 
Resources: 2ùë•+ùë¶‚â§1000
Labour Hours: 3ùë•+2ùë¶‚â§600
Non-negativity: ùë•,ùë¶‚â•0

Problem Setup

```python
# Importin libraries
import pulp

# create a linear Programming problem
model = pulp.LpProblem("Maximize_Profit", pulp.LpMaximize)

# Decision Variable 
x = pulp.LpVariable('Product_A', lowBound=0, cat='Continuous')
y = pulp.LpVariable('Product_B', lowBound=0, cat='Continuous')

# Objective Function
model += 50 * x + 30 * y, "Profit"

# Constraints
model += 2 * x + y <= 1000, "Resource_Constraint" 
model += 3 * x + 2 * y <= 600, "Labour_Constraint"
```

Solving the Problem

```python
# Solve the problem
model.solve()

# Output the results
print(f"Status: {pulp.LpStatus[model.status]}")
print(f"Optimal number of Product A to produce: {pulp.value(x)}")
print(f"Optimal number of Product B to produce: {pulp.value(y)}")
print(f"Maximum Profit: {pulp.value(model.objective)}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

x_vals = np.linspace(0, 500, 500)
y1_vals = 1000 - 2 * x_vals
y2_vals = (600 - 3 * x_vals) / 2
y3_vals = np.maximum(0, np.minimum(y1_vals, y2_vals))

plt.plot(x_vals, y1_vals, label=r'$2x + y \leq 1000$')
plt.plot(x_vals, y2_vals, label=r'$3x + 2y \leq 600$') 
plt.fill_between(x_vals, 0, y3_vals, alpha=0.3) 
plt.xlabel('Product A')
plt.ylabel('Product B') 
plt.title('Feasible Region')
plt.legend() 
plt.show()
```
